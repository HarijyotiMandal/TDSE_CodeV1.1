```python
#Electric Field
# Atomic Unit Conversion
au_time = 2.4188843265864e-2 # fs
hartree_energy = 27.211386245981 # eV
au_time_conversion = 1 / au_time # from fs to au_time
au_energy_conversion = 1 / hartree_energy # from eV to au_energy
def C_t(t, t_min, t_max, delta_T):
    return np.where(
        t < t_min - delta_T, 0,
        np.where(t < t_min, np.sin((t - (t_min - delta_T)) * (np.pi / (2 * delta_T))) ** 2,
        np.where(t <= t_max, 1,
        np.where(t <= t_max + delta_T, np.cos((t - t_max) * (np.pi / (2 * delta_T))) ** 2, 0))))

# Function for E_omega (fundamental frequency)
def E_omega(t, E0, omega, t0, tau, phi_w):
    envelope = np.exp(-((t - t0) / tau) ** 2)
    oscillation_w = np.cos(omega * t + phi_w)
    return (E0 * envelope * oscillation_w )* C_t(t, t_min, t_max, delta_T)

# Time array(in fs)
start_time, stop_time = -200,200  # fs
steps = 40000  # steps
t_values_SI = np.linspace(start_time, stop_time, steps)  # Time array in femtoseconds
dt_SI = t_values_SI[1]- t_values_SI[0]
print(f'Time Step is {np.round(dt_SI*1000, decimals =2)} attoseconds')
t_values = t_values_SI *au_time_conversion  # Time array in atomic units
# Time step
dt = t_values[1] - t_values[0]
# Electric Field parameters
phi_w = 0.0
tau = 50  # fs
tau_au = tau * au_time_conversion
t0 = 0 # fs  # setting t0 to a non-zero value to avoid zero limit issue
t0_au = t0 * au_time_conversion
E0_au = 0.00005
E02_au = 0.000005
omega = 21.2*au_energy_conversion

# Markers and annotations
t_min = -(tau-5)*au_time_conversion*3
t_max = (tau-5)*au_time_conversion*3
delta_T = 4*au_time_conversion

E_om = E_omega(t_values, E0_au, omega, t0_au, tau_au, phi_w)
```

```python
# Initialize lists to store results
results = []

# Parameters
steps_eV = 0.001
omega_values = np.round(np.arange(21.1, 21.3, steps_eV), 3)
iloc_val = 33900
phi_values = np.round(np.arange(0, 2 * np.pi, steps_phi), 7)

# Loop over phi_values and omega_values
for phase in phi_values:
    for omega_ev in omega_values:
        # Construct the key for the DataFrame
        key = f'df_psi_{omega_ev}_{phase}'
        if key in results_dfs:
            # Get the DataFrame
            df = results_dfs[key]
            
            # Calculate populations
            population_s = np.abs(df['s'].iloc[iloc_val])**2 
            population_p = np.abs(df['p'].iloc[iloc_val])**2
            population_es = np.abs(df['es'].iloc[iloc_val])**2
            population_ep = np.abs(df['ep'].iloc[iloc_val])**2
            population_ed = np.abs(df['ed'].iloc[iloc_val])**2

            amp_s = np.abs(df['s'].iloc[iloc_val])
            amp_p = np.abs(df['p'].iloc[iloc_val])
            amp_es = np.abs(df['es'].iloc[iloc_val])
            amp_ep = np.abs(df['ep'].iloc[iloc_val])
            amp_ed = np.abs(df['ed'].iloc[iloc_val])


            
            phase_es = np.angle(df['es'].iloc[iloc_val])
            phase_ep = np.angle(df['ep'].iloc[iloc_val])
            phase_ed = np.angle(df['ed'].iloc[iloc_val])

            # Transform to range [0, 2) element-wise
            phase_es_0_to_2pi = phase_es if phase_es >= 0 else phase_es + 2 * np.pi
            phase_ep_0_to_2pi = phase_ep if phase_ep >= 0 else phase_ep + 2 * np.pi
            phase_ed_0_to_2pi = phase_ed if phase_ed >= 0 else phase_ed + 2 * np.pi
            
            # Append results to list
            results.append({
                'omega_ev': omega_ev,
                'phase': phase,
                'population_s': population_s,
                'population_p': population_p,
                'c0': amp_es,
                'c1': amp_ep,
                'c2': amp_ed, 
                'phi0': phase_es_0_to_2pi,
                'phi1': phase_ep_0_to_2pi,
                'phi2': phase_ed_0_to_2pi
            })
        else:
            print(f"Key {key} not found in results_dfs.")

# Convert results to a DataFrame for easier analysis and saving
results_df = pd.DataFrame(results)

# Save to a CSV file
results_df.to_csv('populations_and_phases_V2.csv', index=False)

print("Populations and phases have been saved to 'populations_and_phases.csv'.")
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from scipy.special import sph_harm_y
%matplotlib widget
# Function to compute spherical harmonics
def compute_spherical_harmonics(l, m, theta, phi):
    return sph_harm_y(l, m, theta, phi)
# Generate the grid
theta = np.linspace(0, np.pi, 100)
phi = np.linspace(0, 2 * np.pi, 100)
theta, phi = np.meshgrid(theta, phi)

# Compute spherical harmonics
Y00 = compute_spherical_harmonics(0, 0, theta, phi)
Y10 = compute_spherical_harmonics(1, 0, theta, phi)
Y20 = compute_spherical_harmonics(2, 0, theta, phi)




# Define the coefficients
C0 = 1.0
C1 = 0.01
C2 = 1.0
# Define the phase shifts
phi0 = 0.0
phi1 = np.pi / 4
phi2 = 0.0
# Define the phase difference
del_Phi = phase_diff
# Compute the spherical harmonics with coefficients and phase shifts
def superposition_spherical_harmonics(C0, C1, C2, Y00, Y10, Y20, phi0, phi1, phi2, del_Phi):
    P = C0 * Y00 * np.exp(1j * phi0) + C1 * Y10 * np.exp(1j * (phi1 + del_Phi)) + C2 * Y20 * np.exp(1j * phi2)
    return P

# Convert to Cartesian coordinates for plotting
def spherical_to_cartesian(r, theta, phi):
    x = r * np.sin(theta) * np.cos(phi)
    y = r * np.sin(theta) * np.sin(phi)
    z = r * np.cos(theta)
    return x, y, z
x00, y00, z00 = spherical_to_cartesian(np.abs(Y00), theta, phi)
x10, y10, z10 = spherical_to_cartesian(np.abs(Y10), theta, phi)
x20, y20, z20 = spherical_to_cartesian(np.abs(Y20), theta, phi)

P_Array = []
xp_Array = []
yp_Array = []
zp_Array = []
for index in df.index:
    # Define the coefficients
    C0 = df['c0'].loc[index]
    C1 = df['c1'].loc[index]
    C2 = df['c2'].loc[index]
    # Define the phase shifts
    phi0 = df['phi0'].loc[index]
    phi1 = df['phi1'].loc[index]
    phi2 = df['phi2'].loc[index]
    # Define the phase difference
    del_Phi = 6.28
    # Compute the superposition of spherical harmonics
    P = superposition_spherical_harmonics(C0, C1, C2, Y00, Y10, Y20, phi0, phi1, phi2, del_Phi)
    P_Array.append(P)
    xp, yp, zp = spherical_to_cartesian(np.abs(P), theta, phi)
    xp_Array.append(xp)
    yp_Array.append(yp)
    zp_Array.append(zp)
```

```python
%matplotlib ipympl
# Electric field functions
def E_omega(t, E0_au, omega_au, t0_au, tau_au, phi1):
    # Set the time limit to 3 standard deviations (tau_au)
    t_limit = 3 * tau_au
    time_abs = np.abs(t - t0_au)
    
    return np.where(time_abs < t_limit, 
                    E0_au * np.exp(-((t - t0_au) / tau_au) ** 2) * ((np.cos(omega_au * t + phi1))**2), 
                    0)

def E_2omega(t, E02_au, omega_au, t0_au, tau_au, phi2):
    t_limit = 3 * tau_au
    time_abs = np.abs(t - t0_au)
    return np.where(time_abs < t_limit, 
                    E02_au * np.exp(-((t - t0_au) / tau_au) ** 2) * ((np.cos(2*omega_au * t + phi2))**2),
                    0)

# Electric Field parameters
phi1 = 0.0
phi2 = 0.0
tau = 50  # fs
tau_au = tau * 41
t0 = 0 # fs  # setting t0 to a non-zero value to avoid zero limit issue
t0_au = t0 * 41
E0_au = 0.0005
E02_au = 0.00005
omega_au = 21.2
E_om = E_omega(t_values, E0_au, omega_au, t0_au, tau_au, phi1)
E_2om = E_2omega(t_values, E02_au, omega_au, t0_au, tau_au, phi2)
plt.figure()
plt.plot(t_values_SI, E_om)
plt.plot(t_values_SI, E_2om)
plt.show()
```
